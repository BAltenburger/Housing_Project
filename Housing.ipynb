```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import numpy as np
# importing data as a data frame
cville = pd.read_csv("All-Transactions-City.csv", parse_dates=["DATE"])
us = pd.read_csv("All-Transactions-USA.csv", parse_dates=["DATE"])

# makes both data sets the same length   
us = us.drop(us.index[0:30])
cville = cville.drop([147, 148, 149, 150, 151])

# sets the prices adujsted to the CPI of 2020
cpi_1980 = 82.4
cpi_1995 = 152.4
cpi_2020 = 258.811
us["USSTHPI"] = us["USSTHPI"] * (cpi_1980 / cpi_2020)
cville["ATNHPIUS16820Q"] = cville["ATNHPIUS16820Q"] * (cpi_1995 / cpi_2020)

cville = cville.loc[(cville.index >= 106) & (cville.index < 146)]
us = us.loc[(us.index >= 136) & (us.index < 176)]

for i in cville["ATNHPIUS16820Q"]:
    i =  i - np.mean(cville["ATNHPIUS16820Q"])
for i in us["USSTHPI"]:
    i =  i - np.mean(us["USSTHPI"])


plt.hist(cville["ATNHPIUS16820Q"], bins = 5, alpha = 0.5, label = "Charlottesville")
plt.hist(us["USSTHPI"], bins = 5, alpha = 0.5, label = "US")
plt.xlabel("House Price Index 2020=100")
plt.ylabel("Frequency")
plt.legend(loc="upper right")
plt.title("Housing Price Index adjusted for 2020 CPI from 2010 to Present")
plt.show()
```
